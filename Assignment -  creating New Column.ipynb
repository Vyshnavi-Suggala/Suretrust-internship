```python
import pandas as pd  #used to handle data
import matplotlib.pyplot as plt #used to create graph
```

```python
data = {
    "User Name": ["Asha", "Ravi", "Priya", "Kiran", "Sneha", "Arjun", "Meena", "Rahul", "Divya", "Vijay"],
    "Duration": [240, 180, 210, 300, 260, 150, 220, 200, 280, 240],
    "Listening Time": [200, 50, 190, 120, 250, 60, 180, 30, 260, 100]
} # Creating Dataset
```

```python
df = pd.DataFrame(data) #Converting into Table Format For Understanding
```

```python
print(df)
```

```python
df["Listening %"] = (df["Listening Time"] / df["Duration"]) * 100  #Creating New Column
 # Listening :How many time the user listened song
 # why we added : TO know the user interest
```

```python
print(df)
```

```python
plt.figure()
plt.bar(df["User Name"], df["Listening %"])

plt.xlabel("User Name")
plt.ylabel("Listening Percentage")
plt.title("User Interest Level Based on Listening %")

```

```python
def interest_level(percent):
    if percent > 80:
        return "High"
    elif percent >= 40:
        return "Medium"
    else:
        return "Low"

df["Interest Level"] = df["Listening %"].apply(interest_level)
```
